```python
import matplotlib.pyplot as plt
import numpy as np
import math
a = int(input("Enter a number: "))
b = 0
alist = []
alog = []
blist = []
while a != 1 :
    if a%2 ==0:
        a = a/2
        #print(a)
        alist.append(a)
        alogint = math.log(a,10)
        alog.append(alogint)
    else:
        a = 3*a+1
        alist.append(a)
        alogint = math.log(a,10)
        alog.append(alogint)
    
    b= b+1
    blist.append(b)

#print(b)

alognum = np.array(alog)
bnum = np.array(blist)
alistnum = np.array(alist)

plot1 = plt.figure(1)
plt.plot(bnum,alistnum)
plot2 = plt.figure(2)
plt.plot(bnum,alognum)
plt.show
```
